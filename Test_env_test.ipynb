```python
import requests
import json

class Instant:
    def __init__(self):
        self.base_url = "https://temis.taipeifubon.com.tw/uma_instant_api"
        self.log_in = {"identifier": "youhsuan.yen",
                       "password": "Aa87522899"}

        self.log_in_headers = {
            'User-Agent': 'PostmanRuntime/7.28.0',
            'Content-Type': 'application/json'
        }
        # get token
        api_endpoint = self.base_url + "/api/auth/authenticate"
        r = requests.post(url=api_endpoint, data=json.dumps(self.log_in), headers=self.log_in_headers).json()
        print(r)
        token = r['data']['token']
        print(token)
        access_token = "Bearer " + token
        # new headers
        self.headers = {
            'User-Agent': 'PostmanRuntime/7.28.0',
            'Content-Type': 'application/json',
            "Authorization": access_token
        }

    # 查詢_分類
    def get_category(self):
        api_endpoint = self.base_url + "/api/oppt-category?pageIndex=0&pageSize=1000"
        # get
        data = requests.get(url=api_endpoint, headers=self.headers).json()
        for i in range(len(data['data'])):
            category_id = data['data'][i]['id']
            category_name = data['data'][i]['name']
            print(f'{category_id} : {category_name}')

    # 查詢_圖表
    def get_chart(self):
        api_endpoint = self.base_url + "/api/chart?pageIndex=0&pageSize=1000"
        # get
        data = requests.get(url=api_endpoint, headers=self.headers).json()
        for i in range(len(data['data'])):
            chart_id = data['data'][i]['id']
            chart_name = data['data'][i]['name']
            main_category = data['data'][i]['categoryId']
            sub_category = data['data'][i]['subCategoryId']
            main = "關鍵圖表" if main_category == "1" else "客群解析"
            if sub_category == "2":
                sub = "人口統計"
            elif sub_category == "4":
                sub = "行為指標"
            elif sub_category == "5":
                sub = "地理關係"
            elif sub_category == "6":
                sub = "心理圖"
            else:
                sub = "無次分類"
            print(f'{chart_id}: {chart_name} ; {main}>{sub}')

    # 查詢_商機框架
    def get_template(self):
        #https://api.dev.instant.profyu.com/api/oppt-template?pageIndex=0&pageSize=100
        api_endpoint = self.base_url + "/api/oppt-template?pageIndex=0&pageSize=1000"
        # get
        r = requests.get(url=api_endpoint, headers=self.headers).json()
        print(json.dumps(r, indent=4, sort_keys=True, ensure_ascii=False))

    # 新增_商機分類
    def post_category(self, category_name):
        data = {
            'name': category_name
        }

        api_endpoint = self.base_url + "/api/oppt-category"

        # post
        r = requests.post(url=api_endpoint, data=json.dumps(data), headers=self.headers).json()
        category_id = r['id']
        category_new_name = r['name']
        print(f'{category_id} : {category_new_name}')

    # 新增_圖表
    def post_chart(self, chart_data):
        # categoryId（主分類）可以選擇：
        # （1）關鍵圖表
        # （3）客群解析
        # subCategoryId（次分類）可以選擇：
        # （2）人口統計
        # （4）行為指標
        # （5）地理關係
        # （6）心理圖
        # （0）無次分類

        # chartNo要不同
        # name 是圖表名稱
        # chartDesc 是圖表敘述

        api_endpoint = self.base_url + "/api/chart"
        # post
        r = requests.post(url=api_endpoint, data=json.dumps(chart_data), headers=self.headers).json()
        print(r)
        chart_id = r['id']
        chart_name = r['name']
        print(f'{chart_id} : {chart_name}')

    # 新增_商機框架
    def post_template(self, template_data):
        # templateNo 為商機偵測表裡的ID
        # startTime & endTime 為設定時間 +8hr = 現在時間
        # categoryID 要從get_category裡面選
        # chartId 要從get_chart裡面選
        
        api_endpoint = self.base_url + "/api/oppt-template"
        # post
        r = requests.post(url=api_endpoint, data=json.dumps(template_data), headers=self.headers).json()
        print(r)
        template_id = r['id']
        template_name = r['name']
        print(f'{template_id} : {template_name}')
    
    #停止_偵測
    def stop_detect_template(self, template_id):
#         https://temis.taipeifubon.com.tw/uma_instant_api/api/oppt-template/8171251580954079232/enabled?enabled=false
        endpointFirst = self.base_url + "/api/oppt-template/"
        endpointSec = "/enabled?enabled=false"
        api_endpoint = endpointFirst + template_id + endpointSec
        r = requests.post(url=api_endpoint, headers=self.headers).json()
        print(r)
    #啟動_偵測
    def start_detect_template(self, template_id):
#         https://temis.taipeifubon.com.tw/uma_instant_api/api/oppt-template/8171251580954079232/enabled?enabled=false
        endpointFirst = self.base_url + "/api/oppt-template/"
        endpointSec = "/enabled?enabled=true"
        api_endpoint = endpointFirst + template_id + endpointSec
        r = requests.post(url=api_endpoint, headers=self.headers).json()
        print(r)   
# 後端A <<< API << 後台介面(現在沒有)，用PYTHON代替 
# API >>>> 後端A >>> API >> 前端Instant 介面        
```
